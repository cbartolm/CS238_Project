### Import Libraries 

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns

from matplotlib import style
style.use('ggplot')
```

### Bumper Baseline 

```python
bumper_standard = pd.read_csv("./baseline_bumper.csv")
plt.figure(1, figsize=(8,6))

plt.plot(bumper_standard.iloc[:,0],bumper_standard.iloc[:,1])
plt.title("Total Reward of the different experiments for the bumper baseline")
plt.ylabel("Total Reward")
plt.xlabel("Experience number")

# Plot mean value
mean_value = np.mean(bumper_standard.iloc[:,1])
plt.axhline(y=mean_value, xmin=0, xmax=1, linewidth=0.5, color = 'g')

plt.show()

savefig('bumper_curve.png')
```

```python
## v_noise_coefficient plots

y = [-13.94, -13.94, -13.94]
yerr = 2.126

plt.figure(2, figsize=(12,6))
plt.errorbar(x = [0.2, 2, 8], y = y, yerr = yerr, fmt='--o')
plt.title("Mean Total Rewards and Errors for the bumper baseline for different v_noise")
plt.xlabel("v_noise")
plt.ylabel("Mean Total Reward and its error")

savefig('bumper_v_noise.png')
```

```python
## om_noise_coefficient plots

y = [-13.94, -13.94, -13.94]
yerr = 2.126

plt.figure(3, figsize=(12,6))
plt.errorbar(x = [0.2, 0.5, 2.5], y = y, yerr = yerr, fmt='--o')
plt.title("Mean Total Rewards and Errors for the bumper baseline for different om_noise")
plt.xlabel("om_noise")
plt.ylabel("Mean Total Reward and its error")

savefig('bumper_om_noise.png')
```
